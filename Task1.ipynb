```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('E:\\downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_3401680\\API_SP.POP.TOTL_DS2_en_csv_v2_3401680.csv', skiprows=4)
print(df.head())

# Filter for the latest year available (e.g., 2019)
latest_year = '2019'  
if latest_year in df.columns:
    df_latest = df[['Country Name', latest_year]]
    df_latest = df_latest.rename(columns={latest_year: 'Population'})
else:
    raise ValueError(f"Year {latest_year} is not available in the dataset.")

# Drop rows with missing population data
df_latest = df_latest.dropna()

# Sort the data for better visualization
df_latest = df_latest.sort_values(by='Population', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 8))
plt.bar(df_latest['Country Name'], df_latest['Population'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Total Population by Country ({latest_year})')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
